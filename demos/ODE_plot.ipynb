```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot test_ode.cc (Implicit Euler Method)

data = np.loadtxt('test_ode_ie.txt')

plt.plot(data[:,0],data[:,1],label='y1')
plt.plot(data[:,0],data[:,2],label='y2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```

```python
# plot test_newmark.cc (Newmark Method)

data = np.loadtxt('test_ode_newmark.txt')

plt.plot(data[:,0],data[:,1])
plt.xlabel('t')
plt.ylabel('x(t)')
#plt.legend()
plt.show()
```

```python
# plot test_alpha.cc (Generalized alpha Method)

data = np.loadtxt('test_ode_alpha.txt')

plt.plot(data[:,0],data[:,1],label='x1')
plt.plot(data[:,0],data[:,2],label='x2')
plt.plot(data[:,0],data[:,3],label='x3')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```
