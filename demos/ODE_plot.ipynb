Warning: The test files generate txt files which contain the respective results used in the plots. These txt are saved in your current working directory, so make to compile the test files while you are in "../demos/" (or change absolute paths to your liking).

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot Implicit Euler Method (test_ode.cc)

data = np.loadtxt('results_ode_ie.txt')

plt.plot(data[:,0],data[:,1],label='y1')
plt.plot(data[:,0],data[:,2],label='y2')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.show()
```
